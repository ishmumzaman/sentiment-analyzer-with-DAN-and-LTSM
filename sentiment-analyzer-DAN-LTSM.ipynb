```python

import re
import string
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from gensim.models import Word2Vec
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GlobalAveragePooling1D, Dense, Dropout, Bidirectional, LSTM
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc, ConfusionMatrixDisplay, roc_auc_score
from tensorflow.keras.callbacks import EarlyStopping
from keras.models import Sequential
from keras.layers import Embedding, Bidirectional, LSTM, Dropout, Dense
from keras.callbacks import EarlyStopping


```

```python
#Removing reviews with 1000+ words (as those can be considered outliers)_
df = df[df['review_length'] <= 1000].copy()
removed_count = (df['review_length'] > 1000).sum()
print(f"Removed {removed_count} reviews longer than 1000 words.")

#Seperate function to clean text

def preprocess_text(text):
    # Lowercase
    text = text.lower()
    
    # Remove punctuation
    text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)
    
    # Tokenize by whitespace
    tokens = text.split()
    
    # Remove stopwords
    tokens = [word for word in tokens if word not in ENGLISH_STOP_WORDS]
    
    # Join back to string
    return " ".join(tokens)

df['Review_clean'] = df['Review'].apply(preprocess_text)
df.columns

```

```python
VOCAB_SIZE = 10000
OOV_TOKEN = "<OOV>"
MAX_LEN = 200  # Can adjust based on review length analysis if needed

# Initialize and fit tokenizer
tokenizer = Tokenizer(num_words=VOCAB_SIZE, oov_token=OOV_TOKEN)
tokenizer.fit_on_texts(df['Review_clean'])

# Convert to sequences
sequences = tokenizer.texts_to_sequences(df['Review_clean'])

# Pad sequences
padded_sequences = pad_sequences(sequences, maxlen=MAX_LEN, padding='post', truncating='post')
labels = df['Positive Review'].values
X_temp, X_test, y_temp, y_test = train_test_split(
    padded_sequences, labels, test_size=0.2, random_state=42, stratify=labels
)

X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.25, random_state=42, stratify=y_temp
)
```

```python


# Prepare sentences as lists of tokens
sentences = [text.split() for text in df['Review_clean']]

# Train Word2Vec
w2v_model = Word2Vec(
    sentences,
    vector_size=300,    # embedding dim
    window=5,
    min_count=2,        # ignore very rare words
    workers=4,
    epochs=10
)

# Access the keyed vectors
w2v = w2v_model.wv

VOCAB_SIZE = 10000
EMBEDDING_DIM = 300

# word_index from your fitted tokenizer
word_index = tokenizer.word_index

# Initialize matrix with zeros
embedding_matrix = np.zeros((VOCAB_SIZE, EMBEDDING_DIM))

# Fill matrix: for each word in your vocab, if it's in w2v, copy its vector
for word, idx in word_index.items():
    if idx < VOCAB_SIZE:
        if word in w2v:
            embedding_matrix[idx] = w2v[word]
        else:
            # you can leave it as zeros or assign a small random vector
            embedding_matrix[idx] = np.random.normal(scale=0.6, size=(EMBEDDING_DIM,))


##DAN
dan = Sequential([
    Embedding(input_dim=VOCAB_SIZE,
              output_dim=EMBEDDING_DIM,
              weights=[embedding_matrix],
              input_length=MAX_LEN,
              trainable=False),
    GlobalAveragePooling1D(),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

dan.compile(loss='binary_crossentropy',
            optimizer='adam',
            metrics=['accuracy'])
dan.summary()
```

```python
#Training data on the DAN model

history = dan.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=100,
    batch_size=32,
    verbose=1
)
```

```python
# Plot Accuracy
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# Plot Loss
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

Looks like the DAN model does not perform that well. In the model accuracy plot, validation accuracy stays flat (~0.58–0.60), and quite noisy and the gap between train/val accuracy grows, these are signs of the model generalizing not that well and overfitting. In the model loss plot, Validation loss increases overtime, which proves that the model is overfitting. Hence I want to try LTSM

```python



# LSTM Model
lstm_model = Sequential([
    Embedding(input_dim=VOCAB_SIZE,
              output_dim=EMBEDDING_DIM,
              weights=[embedding_matrix],
              input_length=MAX_LEN,
              trainable=False),  
    
    Bidirectional(LSTM(64, return_sequences=False)),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

lstm_model.compile(loss='binary_crossentropy',
                   optimizer='adam',
                   metrics=['accuracy'])

# Print model summary
lstm_model.summary()
history_lstm = lstm_model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=10,           
    batch_size=64,
    verbose=1
)

```

```python
# Accuracy plot
plt.plot(history_lstm.history['accuracy'], label='Train Accuracy (LSTM)')
plt.plot(history_lstm.history['val_accuracy'], label='Val Accuracy (LSTM)')
plt.title('LSTM Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# Loss plot
plt.plot(history_lstm.history['loss'], label='Train Loss (LSTM)')
plt.plot(history_lstm.history['val_loss'], label='Val Loss (LSTM)')
plt.title('LSTM Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python

import seaborn as sns

# Predict class probabilities and convert to binary
y_pred_prob_lstm = lstm_model.predict(X_val)
y_pred_lstm = (y_pred_prob_lstm > 0.5).astype("int32")

# Confusion Matrix
cm = confusion_matrix(y_val, y_pred_lstm)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("LSTM Confusion Matrix")
plt.show()

# Classification Report
print(classification_report(y_val, y_pred_lstm))
```

```python


fpr, tpr, _ = roc_curve(y_val, y_pred_prob_lstm)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, label='ROC Curve (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('LSTM ROC-AUC Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

From the AOC ROC curves and the plot of Model Accuracy and Model loss, we can clearly see the LTSM model is performing better than DAN model. Train vs val gap is smaller than DAN, which suggests less overfitting than DAN. But in the model loss plot we can see validation loss dips, then spikes, which means model is overfitting around epoch 6–8. Performance is better than random (0.50), but not by much. If I try further tuning or hyperparameter optimization, I probably can increase performance.

```python
#Making another LTSM beside the original one to try out different hyperparameters 
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

lstm_tuned = Sequential([
    Embedding(input_dim=VOCAB_SIZE,
              output_dim=EMBEDDING_DIM,
              weights=[embedding_matrix],
              input_length=MAX_LEN,
              trainable=True),  

    Bidirectional(LSTM(128, return_sequences=False)),
    Dropout(0.4),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

lstm_tuned.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

history_tuned = lstm_tuned.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=20,
    batch_size=64,
    callbacks=[early_stop],
    verbose=1
)
```

```python
# Accuracy plot
plt.plot(history_tuned.history['accuracy'], label='Train Accuracy (Tuned)')
plt.plot(history_tuned.history['val_accuracy'], label='Val Accuracy (Tuned)')
plt.title('Tuned LSTM Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Loss plot
plt.plot(history_tuned.history['loss'], label='Train Loss (Tuned)')
plt.plot(history_tuned.history['val_loss'], label='Val Loss (Tuned)')
plt.title('Tuned LSTM Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Get binary predictions from probabilities
y_pred_prob = lstm_tuned.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='viridis')


print("Classification Report:\n")
print(classification_report(y_test, y_pred))

```

```python
# ROC-AUC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
auc = roc_auc_score(y_test, y_pred_prob)

plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}")
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

The model is performing significantly better after several hyperparameter tuning (tried changing learning rates, dropout rates, LTSM units, EPOCs and implementing early stopping to stop training when val_loss stops improving to cancel further overfitting). In the modified model, it trains with data very well and eventually reaches 1.0 training accuracy, but the val_accuracy stops around 0.75 which means the model is still overfitting, but earlystop did its job. Overall this tuned model is significantly better than the original LTSM model and DAN models.

```python
#Summary

loss, accuracy = lstm_tuned.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {accuracy:.4f}")
print(f"Test Loss: {loss:.4f}")
```
